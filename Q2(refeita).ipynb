```python
import numpy as np
from sympy import *
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from IPython.display import display
```

```python
n = 10 # numero de pontos
r = 0.5
k = 450

t = np.linspace(0,5,n)
y = np.linspace(800,1000,n)

T,Y = np.meshgrid(t,y)

x=r*Y-450
ang = np.arctan(x) 
dy = np.cos(ang)
dt = np.sin(ang)

plt.quiver(T,Y,dy,dt,angles='xy')
plt.grid('on')
```

```python

def ed(y,t):
  x=r*y-k
  return x

tt = np.linspace(0,5,10)
y0 = 900
sol = odeint(ed,y0,tt)

plt.plot(tt,sol) # plota a curva
plt.quiver(T,Y,dy,dt) 
plt.grid('on') 
```

```python
plt.figure(figsize = (6,6),dpi=100)
plt.quiver(dy,dt,label='$dv/dt$',scale_units='xy',scale=1)
plt.grid('on')
for y0 in range(800,1050,50):
  sol = odeint(ed,y0, tt)
  plt.plot(tt,sol, label=y0)
plt.legend()
plt.xlabel('t')
plt.ylabel('y')
plt.show()
```
