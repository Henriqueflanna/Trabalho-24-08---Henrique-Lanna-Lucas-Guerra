$$dy/dt=rp-k$$

```python
import numpy as np
from sympy import *
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from IPython.display import display
```

```python
n = 10 # numero de pontos
r = 0.5
k = 450

t = np.linspace(0,10,n)
y = np.linspace(40,60,n)

T,Y = np.meshgrid(t,y)

x=r*p-450
ang = np.arctan(x) 
dy = np.cos(ang)
dt = np.sin(ang)

plt.quiver(T,Y,dy,dt,angles='xy')
plt.grid('on')
```

```python

def ed(y,t):
  x=r*p-k
  return x

tt = np.linspace(0,10,10)
y0 = 0
sol = odeint(ed,y0,tt)
plt.plot(tt,sol) # plota a curva
plt.quiver(T,Y,dy,dt)
plt.grid('on')
```

```python
plt.figure(figsize = (6,6),dpi=100)
plt.quiver(T,Y,dy,dt,angles='xy',label='$dv/dt$')
plt.grid('on')
for y0 in range(0,10,2):
  sol = odeint(ed,y0, tt)
  plt.plot(tt,sol, label=y0)
plt.legend()
plt.xlabel('t')
plt.ylabel('y')
plt.show()
```

```python
cs = [1, 15, 0.5, 7, 8]
x = sp.Symbol("x")
f = sp.Function("f")(x)
ed = sp.Eq(f.diff(),0.5*f-450)

sol = sp.dsolve(ed,f)
fd = sol.rhs

c1, _ = tuple(fd.free_symbols)

for i in cs:
  s = fd.subs(c1,i)
  sp.plot(s, show=True)
```
